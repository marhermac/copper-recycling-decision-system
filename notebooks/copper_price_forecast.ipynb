# Copper Price Forecast Analysis

## Business Question

How could copper prices evolve during the next six months?

## Objective

Develop scenario-based projections to support decision-making for urban copper recyclers.

## Forecast Scenarios

Three scenarios are considered:

* Pessimistic
* Conservative
* Optimistic

The objective is not to predict the exact future price, but to evaluate potential decision outcomes under different market conditions.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/raw/copper_forecast.csv")

df
```

```python
plt.figure(figsize=(10,6))

plt.plot(
    df["Mes"],
    df["Pesimista"],
    marker="o",
    label="Pesimista"
)

plt.plot(
    df["Mes"],
    df["Conservador"],
    marker="o",
    label="Conservador"
)

plt.plot(
    df["Mes"],
    df["Optimista"],
    marker="o",
    label="Optimista"
)

plt.title(
    "Copper Price Forecast Scenarios"
)

plt.xlabel("Months")
plt.ylabel("Price (ARS/kg)")

plt.legend()
plt.grid(True)

plt.show()
```

```python
df["Kg_Acumulados"] = df["Mes"] * 15

df["Ingreso_Pesimista"] = (
    df["Kg_Acumulados"]
    * df["Pesimista"]
)

df["Ingreso_Conservador"] = (
    df["Kg_Acumulados"]
    * df["Conservador"]
)

df["Ingreso_Optimista"] = (
    df["Kg_Acumulados"]
    * df["Optimista"]
)

df
```

```python
plt.figure(figsize=(10,6))

plt.plot(
    df["Mes"],
    df["Ingreso_Pesimista"],
    marker="o",
    label="Pesimista"
)

plt.plot(
    df["Mes"],
    df["Ingreso_Conservador"],
    marker="o",
    label="Conservador"
)

plt.plot(
    df["Mes"],
    df["Ingreso_Optimista"],
    marker="o",
    label="Optimista"
)

plt.title(
    "Revenue Forecast by Scenario"
)

plt.xlabel("Months")
plt.ylabel("Revenue (ARS)")

plt.legend()
plt.grid(True)

plt.show()
```

# Conclusions

The scenario analysis suggests that inventory accumulation may significantly increase profitability when copper prices remain stable or increase.

Key findings:

* Immediate sales reduce market risk.
* Inventory accumulation increases exposure to price fluctuations.
* Under the conservative scenario, delaying sales may generate higher revenues.
* Decision-making should balance expected market conditions, storage capacity, and liquidity requirements.

## Final Business Question

> Should the recycler sell today or wait?

The answer depends on the expected market scenario and the recycler's risk tolerance.
