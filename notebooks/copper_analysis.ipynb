# Exploratory Data Analysis (EDA)

## Project

**Copper Recycling Decision System**

### Business Question

> How profitable is the copper recycling business in Argentina, and when is the optimal time to sell accumulated copper?

### Objective of this analysis

Analyze the historical evolution of recycled copper prices in Argentina and identify market trends that may influence selling decisions.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/raw/copper_prices_argentina.csv")

df
```

```python
df.describe()
```

```python
df["Crecimiento_%"] = (
    df["Precio_Cobre_1ra_ARS"]
    .pct_change()
    * 100
)

df
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    df["Año"],
    df["Precio_Cobre_1ra_ARS"],
    marker="o"
)

plt.title("Copper Price Evolution in Argentina")
plt.xlabel("Year")
plt.ylabel("ARS per Kg")

plt.grid(True)

plt.show()
```

## Preliminary Findings

The analysis shows a strong upward trend in recycled copper prices in Argentina.

Between 2024 and 2025, prices increased by approximately 85%.

Between 2025 and 2026, prices increased by approximately 98%.

These findings suggest that both international copper prices and local economic conditions may significantly impact the profitability of urban copper recycling.
