# Recycler Profitability Analysis

## Business Question

Which recovered material generates the highest value for an urban copper recycler?

## Objective

Analyze the contribution of different sources of recovered copper to total monthly income.

This analysis is based on field observations of informal urban recycling activities in Argentina.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/raw/recycler_production.csv")

df
```

```python
precio = 15500

df["Ingreso_ARS"] = df["Cobre_kg"] * precio

df
```

```python
df.sort_values(
    by="Ingreso_ARS",
    ascending=False
)
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    df["Material"],
    df["Ingreso_ARS"]
)

plt.xticks(rotation=45)

plt.title("Monthly Revenue by Material")
plt.ylabel("ARS")

plt.show()
```

## Preliminary Findings

The analysis indicates that stripped copper cable generates the highest monthly revenue among the evaluated materials.

Transformers from microwave ovens also represent a significant source of recovered copper.

These findings suggest that the profitability of urban recycling depends not only on the amount of material collected, but also on the type and quality of copper recovered.
