# Copper Supply Chain Analysis

## Business Question

Who captures the highest economic value in the copper recycling supply chain?

## Objective

Analyze the profit distribution across the copper recycling ecosystem, from urban recyclers to industrial foundries.

## Supply Chain

Urban Recycler
↓
Scrap Dealer
↓
Wholesaler
↓
Foundry



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/raw/copper_supply_chain.csv")

df
```

```python
df["Ganancia_ARS"] = (
    df["Venta_ARS"]
    - df["Compra_ARS"]
)

df
```

```python
ganancia_total = df["Ganancia_ARS"].sum()

df["Participacion_%"] = (
    df["Ganancia_ARS"]
    / ganancia_total
    * 100
)

df
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    df["Actor"],
    df["Ganancia_ARS"]
)

plt.title("Profit Distribution Across the Copper Supply Chain")
plt.ylabel("ARS")

plt.show()
```

# Findings

The analysis indicates that value capture is uneven across the copper recycling supply chain.

Key observations:

* Urban recyclers create the initial economic value through collection and material recovery.
* Scrap dealers and wholesalers capture additional margins through aggregation and logistics.
* Industrial processing stages may capture the highest value due to scale and transformation capabilities.

These findings suggest that supply chain position significantly impacts profitability in the copper recycling ecosystem.
